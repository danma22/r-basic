# Gráficos con python

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

### Ejercicio 1
A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 1 en el enunciado.

```python
plt.figure(figsize=(10,8))

n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X) 
plt.plot(X, Y + 1, color='blue', alpha=1.00)
plt.fill_between(X,Y+1,1, color = "#008080")

plt.plot(X, Y - 1, color='blue', alpha=1.00)
plt.fill_between(X,Y-1,-1, where = Y-1 >= -1, facecolor="#008080", interpolate=True)
plt.fill_between(X,Y-1,-1, where = Y-1 <= -1, facecolor="#FC5A50", interpolate=True)

plt.show()
```

### Ejercicio 2

A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 2 en el enunciado.

```python
n = 1024

plt.figure(figsize=(10,8))
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n) 

ax = plt.scatter(X,Y, c=np.arctan2(Y,X), cmap='hsv')
plt.show()
```

### Ejercicio 3
A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 3 en el enunciado.

```python
plt.figure(figsize=(10,8))

n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white') 
for x, y1, y2 in zip(X, Y1, Y2):     
        plt.text(x, y1 + 0.05, '%.2f' % y1, ha='center', va='bottom', size =14)
        plt.text(x, -y2 - 0.05, '%.2f' % y2, ha = "center", va = 'top', size = 14)
        
plt.ylim(-1.25, +1.25)
plt.show()
```

### Ejercicio 4
A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 4 en el enunciado.

```python
plt.figure(figsize=(10,8))
Z = np.arange(0,20)
Z[:-1] = 100/21
Z[-1] = 200/21

explode = [0.05]*19
explode.append(0.1)
c = np.arange(0,1,0.05)
cls = []
for data in c:
    cls.append(str(data))
print(Z)

plt.pie(Z, colors=cls, 
        explode = explode, 
        labeldistance=1.5,
        wedgeprops=dict(width=0.98,edgecolor='k', linewidth=1))
plt.show()
```
