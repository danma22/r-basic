### Cómo crear un data frame

```python
import numpy as np
```

```python
# Se usa la función DataFrame de pandas
df = pd.DataFrame({
    'A' : np.random.rand(10),
    'B' : np.random.rand(10),
    'C' : np.random.rand(10)
})
```

```python
df
```

```python
df.mean()
```

```python
df.mean(axis="columns")
```

```python
df2 = pd.DataFrame({
    "key" : ['A', 'B', 'C', 'A', 'B', 'C'],
    'values' : range(6)
})

df2
```

```python
# Groupby es como aggregate en R, de una columna obtiene ciertas agrupaciones
# Cada agrupación son los valores distintos que hay. Se puede operar con esto
df2.groupby('key')
```

```python
df2.groupby('key').mean() # Agrupar la tabla de acuerdo a los valores del parametro dado
```

### Manipulación de data frames

```python
df = pd.DataFrame({
    'key' : ['A','B','C','A','B','C'],
    'value1' : range(6),
    'value2' : np.random.randint(0,100,6)
})
```

```python
df
```

### Aggregate

```python
# Para aplicar con aggregate las funciones queridas sobre ciertos parametros
# Le pasamos como argumento un vector, lista o string con las fnciones queridas
df.groupby('key').aggregate(['min', np.median, np.mean, max])
```

```python
# En un df agrupado, aplicando una sola función a cada columna 
df.groupby('key').aggregate({
    'value1' : "min",
    'value2' : 'max'
})
```

```python
# Aplicando aggregate a todo el df, en este caso los resultados de cada función
# se aplican a todos los datos y por ende se muestran las funciones en fila
df.aggregate([np.mean, np.median])
```

### Filter

```python
# sample standar derivation para cada grupo de datos según key
df.groupby('key').std()
```

```python
def filter_func(x):
    return x['value2'].std() > 53
```

```python
# La función filter acepta como parametro otra función que servirá como
# un conjunto de condiciones para filtrar los datos
df.groupby('key').filter(filter_func)
```

### Transform

```python
df.groupby('key').transform(lambda x: x - x.mean())
```

```python
df['value1'].transform([np.sqrt, np.exp])
```

### Apply

```python
def norm_by_col2(x):
    x['value1'] /= x['value2'].sum()
    return x
```

```python
df.groupby('key').apply(norm_by_col2)
```

```python
df.apply(lambda x: x.sum())
```

### División de datos por columnas

```python
L = [0,1,0,1,2,0]
print(df)
df.groupby(L).sum()
```

```python
# Indicar el parametro por el cual agruparlos
df.groupby('key').sum()
```

```python
# Indicar un vector de los valores de un parametro
df.groupby(df['key']).sum()
```

```python
df2 = df.set_index('key')
print(df2)
```

```python
mapping = {
    'A': 'vowel',
    'B': 'consonant',
    'C': 'consonant'
}
```

```python
df2.groupby(mapping).sum()
```

```python
df2.groupby(str.lower).mean()
```

```python
df2.groupby([str.lower, mapping]).mean()
```

### Gráficos desde data frames y exportación

```python
years = [year for year in range(1900,2020)]
deads = [(y + np.random.uniform(0,100) - 1850) for y in years]
```

```python
df = pd.DataFrame({
    'year': years,
    'deads': deads
})
```

```python
df.head()
```

```python
df.plot(x='year', y='deads', figsize=(10,8))
```
